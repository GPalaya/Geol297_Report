```python
#Importing modules

import pandas as pd
import numpy as np
import seaborn as sns
import ternary
from matplotlib import pyplot as plt

```

```python
fp = (r"E:/NIGS Class Files/Geol 215/Data/Fieldsheet template.csv")

data=pd.read_csv(fp, skiprows=[0],
                usecols=[r'Sample no. ','L', 'I','S','Phi','MPS','OPI','ER','FL','Roundness','I/L','S/L','(L-I/L-S)'])
```

```python
df = pd.DataFrame(data)

# Calculate statistics for each column
stats = df.agg({
    'MPS': ['mean','std','max','min'],
    'OPI': ['mean','std','max','min'],
    'ER' : ['mean','std','max','min'],
    'FL' : ['mean','std','max','min']
})
stats= stats.round(2)
stats_df = pd.DataFrame(stats)

print(stats_df)
```

```python
# Calculating MPS, FL, and OPI
df = pd.DataFrame(data)

mean_mps = df['MPS'].mean().round(2)
mean_fl = df['FL'].mean().round(2)
mean_opi = df['OPI'].mean().round(2)


#Create dataframe for mean
mean_df = pd.DataFrame({
    'Mean MPS': [mean_mps],
    'Mean FL': [mean_fl],
    'Mean OPI': [mean_opi]
})


# Define a function to determine environment based on specific criteria
def determine_environment(row):
    if (row['Mean MPS'] < 0.65) and (row['Mean FL'] > 45):
        return 'Fluvial'
    elif (row['Mean MPS'] < 0.65) and (row['Mean FL'] <= 45):
        return 'Beach'
    else:
        return 'Unknown'

# Apply the function to each row in the DataFrame to create a new column for the environment
mean_df['Depositional Environment'] = mean_df.apply(determine_environment, axis=1)

# Output the relevant columns
print(mean_df[['Mean MPS', 'Mean FL', 'Mean OPI', 'Depositional Environment']])

```

```python
# Set Seaborn theme, context, and style
sns.set_theme()  # Optionally, pass arguments to configure theme
sns.set_context('paper')
sns.set_style('white')

# Assuming df1 is your pebble data and df2 is your mean values DataFrame
df1 = pd.DataFrame(data)  # Convert data to DataFrame if not already
df2 = pd.DataFrame(mean_df)

# Define boundary values for quadrants
mps_boundary = 0.65
FL_boundary = 45

# Plot MPS/FL graph
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# All Pebbles plot on first axis
sns.scatterplot(ax=ax[0], x='MPS', y='FL', data=df1, s=50, color='b', edgecolor='k', legend=False)
ax[0].text(0.3, 20, 'Beach', fontsize=12)
ax[0].text(0.8, 80, 'Fluvial', fontsize=12)


# Mean pebbles plot on second axis
sns.scatterplot(ax=ax[1], x='Mean MPS', y='Mean FL', data=df2, s=50, color='b', edgecolor='k', legend=False, marker='^')
ax[1].text(0.3, 20, 'Beach', fontsize=12)
ax[1].text(0.8, 80, 'Fluvial', fontsize=12)

# Set axis limits for both subplots
ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 100)
ax[1].set_xlim(0, 1)
ax[1].set_ylim(0, 100)

# Draw boundary lines for environment classification on both axes
ax[0].axhline(FL_boundary, color='gray', linestyle='--')
ax[0].axvline(mps_boundary, color='gray', linestyle='--')

ax[1].axhline(FL_boundary, color='gray', linestyle='--')
ax[1].axvline(mps_boundary, color='gray', linestyle='--')

# Label axes
ax[0].set_xlabel("Sphericity", fontsize=12, ha='center')
ax[0].set_ylabel("Coefficient of Flatness", fontsize=12)

ax[1].set_xlabel("Sphericity", fontsize=12, ha='center')
ax[1].set_ylabel("Coefficient of Flatness", fontsize=12)

# Add titles to subplots
fig.suptitle("Coefficient of Flatness vs Maximum Projection Sphericity (Stratten, 1974)", ha='center', fontsize=14)
ax[0].set_title('All Pebble Samples')
ax[1].set_title('Mean Pebble Samples')

#Save plot
plt.savefig('combined_plot', dpi=300)
# Show plot
plt.show()
```

```python
# Set Seaborn theme, context, and style
sns.set_theme()  # Optionally, pass arguments to configure theme
sns.set_context('paper')
sns.set_style('white')

# Zingg diagram
data = pd.DataFrame(data)
num_samples = len(data)  # Define number of samples after converting to DataFrame

# Define boundary values for Zingg diagram
zingg_boundary = 2/3

# Scatterplot for sphericity diagram for all data
plt.figure(figsize=(6, 5))
sns.scatterplot(x='I/L', y='S/L', data=data, s=50, color='b', edgecolor='k', legend=False)

# Set axis limits
plt.xlim(0, 1)
plt.ylim(0, 1)

# Annotate regions with distinct coordinates
plt.text(0.2, 0.9, 'Discoid/Oblate', fontsize=9, ha='center')
plt.text(0.8, 0.9, 'Equant', fontsize=9, ha='center', va='center')
plt.text(0.2, 0.1, 'Bladed', fontsize=9, ha='center', va='center')
plt.text(0.8, 0.1, 'Prolate/Rod', fontsize=9, ha='center', va='center')

# Draw boundary lines for Zingg classification
plt.axhline(zingg_boundary, color='gray', linestyle='--')
plt.axvline(zingg_boundary, color='gray', linestyle='--')

# Label axes and show plot
plt.xlabel("Flatness Ratio (I/L)")
plt.ylabel("Elongation Ratio (S/L)")
plt.title("Pebble Shape (Zingg 1965)")

# Define the counts for each quadrant (replace with actual count logic if needed)
quadrant_counts = {
    'I (Equant)': len(data[(data['I/L'] > zingg_boundary) & (data['S/L'] > zingg_boundary)]),
    'II (Discoid/Oblate)': len(data[(data['I/L'] <= zingg_boundary) & (data['S/L'] > zingg_boundary)]),
    'III (Bladed)': len(data[(data['I/L'] <= zingg_boundary) & (data['S/L'] <= zingg_boundary)]),
    'IV (Prolate/Rod)': len(data[(data['I/L'] > zingg_boundary) & (data['S/L'] <= zingg_boundary)]),
    'Total': len(data)
}

# Convert quadrant_counts to DataFrame for display as table
quadrant_table_df = pd.DataFrame(list(quadrant_counts.items()), columns=['Quadrant', 'Count'])

# Display the counts table below the plot
plt.table(cellText=quadrant_table_df.values, colLabels=quadrant_table_df.columns, cellLoc='center', 
          loc='bottom', bbox=[0, -0.8, 1, 0.5])

plt.subplots_adjust(left=0.2, bottom=0.4)  # Adjust to fit the table below the plot
# Save Plot
plt.savefig('zingg_d', dpi=300)
# Show Plot
plt.show()
```
